### Importando bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Carregando dados

```python
df = pd.read_csv("Position_Salaries.csv")
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

### Treinando e Prevendo usando Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

regressor =  RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X, y)
```

```python
regressor.predict([[6.5]])
```

### Visualizando resultado do Random Forest

```python
X_up = np.arange(X.min(), X.max(), 0.1).reshape(-1, 1)

plt.scatter(X, y, color="red")
plt.plot(X_up, regressor.predict(X_up), color="blue")
plt.xlabel("Nível")
plt.ylabel("Salário")
plt.title("Salário por nível")
plt.show()
```
