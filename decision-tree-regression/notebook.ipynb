### Carregando bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Carregando dados

```python
data = pd.read_csv("Position_Salaries.csv")

X = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
```

### Treinando o modelo de decision tree regression

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

### Prevendo o valor com o modelo

```python
regressor.predict([[6.5]])
```

### Visualizando os resultados

```python
X_up = np.arange(X.min(), X.max(), 0.1)
X_up = X_up.reshape(-1, 1)

plt.scatter(X, y, color="red")
plt.plot(X_up, regressor.predict(X_up), color="blue")
plt.xlabel("Nível")
plt.ylabel("Salário")
plt.title("Salário por nível")
plt.show()
```
