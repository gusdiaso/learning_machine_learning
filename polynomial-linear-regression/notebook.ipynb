### Importando Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Carregando os Dados

```python
df = pd.read_csv("Position_Salaries.csv")
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

### Treinando com Linear Regression

```python
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()
linear_regressor.fit(X, y)
```

### Visualizando os resultados da Regressao Linear

```python
plt.scatter(X, y, color="red")
plt.plot(X, linear_regressor.predict(X), color="blue")
plt.xlabel("Nivel")
plt.ylabel("Salario")
plt.title("Salario x Nivel (Linear Regression)")
plt.show()
```

### Treinando com Polynomial Linear Regression

```python
from sklearn.preprocessing import PolynomialFeatures

polynomial_regressor = PolynomialFeatures(degree=4)
X_poly = polynomial_regressor.fit_transform(X)
linear_regressor_2 = LinearRegression()
linear_regressor_2.fit(X_poly, y)
```

### Visualizando os Resultados da Regressao Polinomial

```python
plt.scatter(X, y, color="red")
plt.plot(X, linear_regressor_2.predict(X_poly), color="blue")
plt.xlabel("Nivel")
plt.ylabel("Salario")
plt.title("Salario x Nivel (Polynomial Regression)")
plt.show()
```

```python
X_up = np.arange(X.min(), X.max(), 0.1).reshape(-1, 1)
X_poly_2 = polynomial_regressor.fit_transform(X_up)

plt.scatter(X, y, color="red")
plt.plot(X_up, linear_regressor_2.predict(X_poly_2), color="blue")
plt.xlabel("Nivel")
plt.ylabel("Salario")
plt.title("Salario x Nivel (Polynomial Regression)")
plt.show()
```

### Prevendo com Regressao Linear

```python
linear_regressor.predict([[6.5]])
```

### Prevendo com Regressao Polinomial

```python
linear_regressor_2.predict(polynomial_regressor.fit_transform([[6.5]]))
```
